```python
import pandas as pd
```

```python
Sales =pd.read_csv(r"D:\DEMO\Sales.csv")
```

```python
Officers= pd.read_csv(r"D:\DEMO\Officers.csv")
```

```python
print(Sales)
```

```python
CH_Sales = Sales[Sales['Order Priority'].isin(['C', 'H','M'])]
```

```python
Country_Sales = (CH_Sales.groupby(['Country', 'Item Type'])['Total Profit'].sum().unstack(fill_value=0))/1000
print(Country_Sales)
```

```python
Country_Sales['Total'] = Country_Sales.sum(axis=1)
print(Country_Sales)
```

```python
print(Officers)
```

```python
Country_Officers = Officers.groupby(['Country', 'Item Type'])['first_name'].count().unstack(fill_value=0)
print(Country_Officers)
```

```python
Country_Officers['Total']= Country_Officers.sum(axis=1)
print(Country_Officers)
```

```python
Country_Sales_reset = Country_Sales.reset_index()
Country_Officers_reset = Country_Officers.reset_index()

# Rename first column to 'Country' if needed
Country_Sales_reset.rename(columns={Country_Sales_reset.columns[0]: 'Country'}, inplace=True)
Country_Officers_reset.rename(columns={Country_Officers_reset.columns[0]: 'Country'}, inplace=True)

merged = pd.merge(
    Country_Sales_reset,
    Country_Officers_reset,
    on='Country',
    how='outer',
    suffixes=('_Profit', '_UsedResources')
)
```

```python
import pandas as pd

# Calculate sum for all numeric columns
grand_total = merged.select_dtypes(include='number').sum()

# Create a DataFrame for the total row, keep the same columns and add a label for Region/Unit
total_row = pd.DataFrame(grand_total).T  # transpose to get 1 row
total_row.index = ['Grand Total']

# If you have a Region or Unit column (non-numeric), add it manually
# For example, if the first column is 'Region' or 'Unit':
total_row['Country'] = 'Grand Total'

# Reorder columns so 'Region' is first (adjust if your label column name is different)
cols = merged.columns.tolist()
cols.remove('Country')  # remove to reinsert at front
cols = ['Country'] + cols
total_row = total_row[cols]

# Append total row to original DataFrame
Report = pd.concat([merged, total_row], ignore_index=False)
```

```python
print(Report)
```

```python
import datetime

# Get today's date in YYYYMMDD format
today = datetime.datetime.now().strftime('%d.%m.%Y')

Report.to_excel(f'D:\\DEMO\\Analysis_Report_{today}.xlsx', index=False)

import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.utils import get_column_letter
import datetime

# Export merged DataFrame to Excel
file_path = f'D:\\DEMO\\Analysis_Report_{today}.xlsx'

# Load workbook and active sheet for styling
wb = load_workbook(file_path)
ws = wb.active

# Define styles
pink_fill = PatternFill(start_color="F2DCDB", fill_type="solid")  # light pink
blue_fill = PatternFill(start_color="D9E1F2", fill_type="solid")  # light blue
bold_font = Font(bold=True)
center_align = Alignment(horizontal="center", vertical="center")
thin_border = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)

# Find the columns in Excel by their headers
headers = [cell.value for cell in ws[1]]

# Apply header formatting: pink fill for '_Profit' columns, blue for '_UsedResources'
for col_idx, header in enumerate(headers, start=1):
    cell = ws.cell(row=1, column=col_idx)
    cell.font = bold_font
    cell.alignment = center_align
    cell.border = thin_border
    
    if header is None:
        continue
    if header.endswith('_Profit'):
        cell.fill = pink_fill
    elif header.endswith('_UsedResources'):
        cell.fill = blue_fill
    else:
        # Normal header color or white fill
        cell.fill = PatternFill(start_color="FFFFFF", fill_type="solid")

# Format all data rows: center alignment, borders, number formatting for numbers
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=ws.max_column):
    for cell in row:
        cell.alignment = center_align
        if isinstance(cell.value, (int, float)):
            cell.number_format = "0.0"
        cell.border = thin_border

# Save the styled file
wb.save(file_path)
```
